# Abstract
Understanding the weather is a very important feature within society.  Modelling and forecasting it is a complex procedure that leverages methods across many disciplines, including mathematics, statistics, and meteorology.  In this analysis, we consider only a small part of the weather, temperature, and attempt to model and forecast it.  We focus on a dataset containing 3 years worth of historical daily average temperatures located in Toronto, Ontario.  We begin with a brief discussion on weather in general and some important facts to consider when modelling.  A preliminary analysis (exploration) of the data is carried out followed by model specification and forecasting. The forecasting period is kept small (55 days) since temperature is notoriously difficult to predict long range.  We observe forecasts in accordance with the general trend of the temperature at the end of the time period and summarize with some concluding remarks.  It is important to keep in mind that temperature is a result of many simultaneously occurring natural phenomena and requires a great deal of data to model properly.  Even though this analysis focuses on general predictions of daily average temperatures it is primarily an exercise in model specification, and should be taken as such.  

# Introduction
The weather is a very important part of modern society whose influence spans nearly every facet of human life.  It has impacts on urban development, the outcome of sporting events, people's mood, transportation, global commerce, the stock market, and national GDP.  Since as early as 600B.C.$^1$ humans have tried to predict the weather using whichever tools they had at their disposal.  In the early stages of weather prediction, astronomical phenomena were recorded and associated with particular weather patterns.  This data would then be used to anticipate the weather after the astronomical events had reappeared.  However, most early techniques were speculative and inherently unreliable.  Accurate weather prediction would require instruments to measure various quantitative features like temperature, precipitation, moisture, among other things$^1$.  As technology improved so did weather data collection, and thus, weather prediction.  In our current age, there is a surplus of publicly and privately available data ranging from air temperature, soil moisture and solar radiation, to wind speeds and humidity$^2$.  In spite of this abundance of data, forecasting the weather accurately and at long ranges is still a great challenge.   

It was not until recently that the weather was discovered to be a chaotic system.  In the second half of the 20$^{th}$ century, mathematician Edward Lorenz$^3$ worked on understanding the behaviour of weather systems.  He noticed that even seemingly basic models were capable of producing extremely chaotic results.   In particular he noticed that these systems generally exhibit nonlinear behaviour with respect to their initial conditions.  This means that the equations defining a chaotic system will produce wildly different results after a slight perturbation in the initial conditions the equations begin at.  This phenomena is sometimes referred to as the Butterfly Effect$^4$ and is beginning to get much more attention in technical disciplines.  The chaotic nature of these systems gives an interesting reason as to why it is so difficult to accurately predict weather at long ranges and is something to keep in mind throughout the analysis.

In our analysis we consider a dataset consisting of 3 years worth of average daily temperatures (degrees celsius) for Toronto, Ontario.  In particular, data was collected for a total of 1095 days ranging from March 18, 2018 to March 16, 2021.  The year 2020 was a leap year which means we take 365.25 days to be in a single year. The frequency of our data (daily) plays an important role within our analysis.  In general, weather in most parts of Canada is seasonal, meaning over a year, temperature is expected to follow a pattern according to the 4 seasons: winter, spring, summer and fall.  While this may be true for our dataset, we do not consider the data to be seasonal in a technical sense.  Broadly speaking, it is not the case that the temperature on a particular day will be indicative of the temperature on the same day a year later.  This is explained in more detail in the section on nonstationarity when we remove it from the time series.  However, we do uncover a deterministic trend and remove it to extract the noise from the series.  This noise is modelled with an $ARMA(3,2)$ model and used to forecast 55 days ahead.  This is reasonable since we expect the period of the deterministic trend to be yearly, of which we have 3 years worth of data, and attempt to forecast 5% of the total dataset.  Likewise, the exploration of the data below is inline with traditional observations and does not appear to experience any major events that must be accounted for.  
 

## Explore Data
To begin our analysis, we start by exploring the data.  The purpose of this exploratory stage is to get an idea of the numerical properties of the data.  In particular, we want to get an idea of the date range, temperature range, uncover any outliers, explore trends, and so on.  First, we import our data:

```r
d <- read.table('Toronto_Temperature.csv', sep=',', header=TRUE) # read data
d$Date <- as.Date(d$Date, format="%Y-%m-%d") # change date format
temps <- d$Temperature  # used later in the analysis
d[, 'y'] <- format(d[, 'Date'], '%Y')
d[, 'm'] <- format(d[, 'Date'], '%m')
d[, 'd'] <- format(d[, 'Date'], '%d')
tail(d)
```

```r
summary(d[, c('Date', 'Temperature')])  # an idea of the range of data
```

```r
plot(x=d[1], y=temps, type='l', xlab='Date', ylab='Temperature (Celcius)', main='Daily Average Temperature')
```

```r
hist(d$Temperature, breaks=20, xlab='Temperature', main='Distribution of Temperatures')
```

As we can see, the data seems clean to begin with.  There appear to be no outliers or missing dates.  Notice from the time series plot that there appears to be a cyclical trend which we explore later.  Since everything seems okay we will start fitting our model. 

# Model Specification
## Confirming Nonstationarity
Here we begin the process of modelling the data.  First we wish to split our dataset into training and validation.  We build a model based on the training data and then evaluate its performance on the test set.  As mentioned earlier, weather is a difficult phenomena to model and forecast errors tend to compound rapidly the further out the forecast.  As such, we forego the typical 90/10 train/test split for a 95/5 split.  This gives us $\lfloor0.95\cdot1095\rfloor=1040$ days worth of data to train our model and $55$ days to forecast and evaluate the model.  

```r
n <- length(temps)
split <- floor(n*0.95) 
train <- d[1:split, ]   
test <- d[(split+1):n, ]
train.temps <- train$Temperature
test.temps <- test$Temperature
```

The first step in any model specification procedure is to consider the ACF and PACF plots of the time series. The slow decay of the ACF indicates an autoregressive aspect to the series and significant PACF values for lags at and before 7 indicate the autoregressive order is around a week.  The PACF values do decay rather quickly so the order might be closer to around 3 days.  This makes sense since the temperature on a particular day is usually indicative of the temperature on the next day (we don't expect to see massive fluctuations from day to day like freezing weather in the middle of the summer or vice versa).  In addition, these plots are another reason to think the initial time series is not stationary, below we confirm these suspicions.  


```r
acf2(train.temps)[0]
```

Here we test whether or not the data is stationary.  Note that the Augmented Dickey-Fuller (ADF) test tests the null hypothesis that the temperatures are not stationary (i.e. presence of a unit root).  We find an insignificant $p$-val of $0.6309$ which means we cannot conclude that the data is stationary.  Likewise, the KPSS test is insignificant for level stationarity but is significant for trend stationarity.  Recall the KPSS test tests the null hypothesis that the data is stationary (opposite to that of ADF) and a significant result allows us to conclude the original time series is not stationary.  In order to follow the Box-Jenkins Methodology, we must remove the nonstationarity.   

```r
# we confirm our hypothesis of nonstationarity with ADF test 
# and KPSS test 
adf.test(train.temps) 
kpss.test(train.temps, null='Level') 
kpss.test(train.temps, null='Trend') 
```

## Removing Notstationarity
We saw above that the time series is not stationary, a requirement for the Box-Jenkins Methodology, so we must remove this nonstationarity.  It was touched upon in the intro that the frequency of the time series is important for analysis.  In particular, we consider daily average temperatures and while weather is usually considered seasonal, this may not be the case on a daily scale.  In a time series, seasonality refers to variables at a particular lag, say $s$, being useful in predicting the time series at the current instance.  For example, if we had monthly temperature data, seasonality would make sense, and variables at lag $s=12$ would be useful in predicting the time series at the current instance (the average temperature in the month of January 2020 will be a good predictor of the average temperature in January 2021).  However, it doesn't seem appropriate to expect that the temperature on a particular day of the year will be a good predictor of that same day in the next year.  That is, the temperature on January 1, 2020 is not expected to be a good predictor of the temperature on January 1, 2021.  Nevertheless, we need to remove the nonstationary portion of the time series.

Although we rule out seasonality, there is still a predictability in the data.  If we refer to the time series plotted above, we notice a clear oscillating pattern, like a sinusoidal wave.  We call this a deterministic trend; a predictable progression of the phenomena that is expected to continue in perpetuity (or at least for the time frame of the analysis).  In our case, we fit a cosine curve to model the deterministic portion of the time series, see Cryer & Chan Time Series Analysis Chapter 3 for more info on the derivation of the regression.  

```r
# In the analysis that follows we make use of time series (ts) objects.
# The start day of the time series can be passed to the object so we 
# must figure out what day of the year the first data point corresponds
# to.
day1 <- train[1, 1] # get first date: March 18, 2018
start.day <- as.numeric(strftime(day1, format = "%j")) 
s <- '18/03/2018 is the {start.day}th day of the year'
print(glue(s))
```

```r
# we can also set frequency.  Since 2020 is a leap year, years are ~365.25 days long.
train.temps.ts <- ts(train.temps, start=c(2018, start.day), f=365.25)  
X <- harmonic(train.temps.ts, 1)    # matrix of predictor variables
x1 <- data.frame(X)$cos.2.pi.t      # predictor 1
x2 <- data.frame(X)$sin.2.pi.t      # predictor 2
cos_model <- lm(train.temps.ts ~ x1 + x2)
summary(cos_model)
cos_model.ts <- ts(fitted(cos_model), start=c(2018, start.day), f=365.25)
plot(cos_model.ts, ylim=range(c(-20, 30)), xlab='Year', ylab='Temperatures', main='Daily Average Temperature with Fitted Cosine Trend')
points(train.temps.ts)
```

As we can see, the fitted cosine curve follows the trend in the temperature data quite nicely.  To extract the stationary portion of the data for the Box-Jenkins Methodology we simply take the difference between the original time series and the deterministic trend.  This new time series will look like white noise.  This is tested using the ADF test like above and we see a significant result, hence we reject the null hypothesis of nonstationarity with confidence (i.e. the new time series is stationary).   

```r
noise <- train.temps.ts - cos_model.ts  # noise = original - deterministic
plot(noise)
adf.test(noise) # pval is significant, hence we reject H0 -> stationary series
```

# Modelling the Noise
Now that we have extracted the nonstationary portion of the data we want to model it.  Similar to the beginning of the analysis we consider the ACF and PACF plots of the noise.  First thing to notice is that the scale on the ACF plot has changed.  This is because we converted our data to a ts object whose scale is in years.  This has no effect on interpreting the plots, just keep in mind that 0.05 years is ~18 days.  We see that the ACF plot decays quickly and the PACF has insignificant spikes for lags greater than 3.  However, neither plot cuts off fully after a particular point.  Instead, they decay slowly, hence we imagine an ARMA model to be best.  We will systematically determine the order of the best model by fitting candidates using the Extended ACF function and comparing diagnostics. 

```r
acf2(noise, max.lag=100)[0] # note that lag is in years, for example 0.05 years is roughly 18 days
```

```r
n_eacf <- eacf(noise, ar.max=5, ma.max=5)
# o represents feasible
# x represent infeasible
```

```r
# evaluate a collection of possible models
noise.aic <- matrix(0, 5, 5)
noise.bic <- matrix(0, 5, 5)
for (i in 0:4) for (j in 0:4){
    noise.fit <- arima(noise, order=c(i, 0, j), method="ML", include.mean = TRUE, optim.control = list(maxit=1000))
    noise.aic[i+1, j+1] <- noise.fit$aic
    noise.bic[i+1, j+1] <- BIC(noise.fit)
}
```

```r
sort(unmatrix(noise.aic, byrow=FALSE))[1:10]
sort(unmatrix(noise.bic, byrow=FALSE))[1:10]
```

# Candidate models
Based on the ACF and PACF and assuming $ARMA(p,q)$ we expect to see $p\geq q$ (otherwise we would see significant lags further out in the PACF).  As such, we consider the following candidates.    
ARMA:
- (4,4)
- (3,3)
- (5,2)
- (5,4)
- (3,2)
- (4,2)

Each model is evaluated along with diagnostics reported in the appendix.  In each case we look at the QQ plot of the residuals, evaluate their diagnostics and test normality using the Shapiro-Wilk test.  Together, these provide a good indication as to whether a models residuals are favorable.  Based on thes results we elect to fit an $ARMA(3,2)$ model.  Specifically, its diagnostics support a normal distribution and it is the most parsimonious of the candidates.  

# Forecasting
## Chosen Model: ARMA(3,2)
Now that we have selected a model, we fit it to the noise and produce forecasts 55 days out.  Recall that the dataset was parsed into training/test using a 95/5 split.  This means we will be forecasting temperatures between January 21, 2021 and March 16, 2021.  Up to this point we have successfully decomposed our time series into two parts; stochastic and deterministic.  The deterministic part is fully determined by its parameters and can be extended easily and with full accuracy.  Since we fit a linear model, we can use the predict function to extend the curve for the 55 days we wish to forecast.  As for the stochastic part, we have fit an $ARMA(3,2)$ model to the noise which we can use to forecast the 55 day period.  For these predictions to be meaningful we simply need to recombine the stochastic part and deterministic part by adding their respective forecasts.  Likewise, error bounds for the $ARMA$ forecast are recorded and reported along with the forecast.  We begin by fitting the model: 

```r
noise.fit <- arima(noise, order=c(3, 0, 2), method='ML', include.mean = TRUE, optim.control = list(maxit=1000))
noise.forecast <- predict(noise.fit, length(test.temps)) # 55 days out
# keep track of error bounds
p_se <- noise.forecast$pred + noise.forecast$se # forecast + (p) standard error (se)
m_se <- noise.forecast$pred - noise.forecast$se # forecast - (m) standard error (se)
```

```r
test_day1 <- test[1, 1] # get first date: January 21, 2021
test.start.day <- as.numeric(strftime(test_day1, format = "%j")) 
s <- '21/01/2021 is the {test.start.day}st day of the year'
print(glue(s))
```

```r
test.temps.ts <- ts(test.temps, start=c(2020, test.start.day), f=365.25)
new_X <- harmonic(test.temps.ts, 1) # produce new set of predictor variables
```

```r
# we need to extend the fitted cosine curve.
# This amounts to predicting new points using
# the linear model found earlier
pred_X <- data.frame(new_X)
colnames(pred_X) <- c("x1", "x2")
preds <- predict(cos_model, pred_X)
```

```r
# extended (predicted) curve in red
plot(fitted(cos_model), 
     xlim=range(c(-1, 1150)), 
     xlab='Time (days)', 
     ylab='Temperature', 
     main='Extended Cosine Curve')
points(c(time(preds) + 1040), preds, col='red')
```

Above we see the extended cosine curve is behaving as expected.  Below we forecast the $ARMA(3,2)$ model 55 days out and recombine with the deterministic portion to obtain the meaningful forecast.  Since our forecast period is between mid January and mid march, we expect to see temperatures around 0 degrees but trending up.  January is usually one of the coldest months of the year but temperatures tend to become positive around March.  We will keep this in mind below

```r
forecast.temps <- noise.forecast$pred + c(preds[1:55])  # recovered forecasts (deterministic + noise)
forecast.p.e <- forecast.temps + noise.forecast$se      # upper (+) error bound
forecast.m.e <- forecast.temps - noise.forecast$se      # lower (-) error bound 
```

```r
# merge train, forecasts and error bounds into single time series
ts.plot(train.temps.ts, 
        forecast.temps, 
        forecast.p.e, 
        forecast.m.e, 
        lty=c(1, 2, 3, 3), 
        col=c(1, 3, 2, 2), 
        ylab='Temperature', 
        main='Daily Average Temperature with Forecast')
```

```r
# Find root mean square error of forecasts
sqrt(sum((c(forecast.temps) + c(test.temps))^2) / length(test.temps))
```

```r
# plotting actual in the time series above is difficult to read
# so here is a zoomed in plot for comparison
ts.plot(test.temps, 
        forecast.temps, 
        forecast.p.e, 
        forecast.m.e, 
        lty=c(1, 2, 3, 3), 
        col=c(1, 3, 2, 2), 
        ylab='Temperature', 
        main='Forecast vs Actual Zoomed In')
```

# Conclusion
Here we evaluate our forecasts and finish with some closing remarks.  The forecasts predict a general upward trend in temperature and we observe a root mean square error of about $9^\circ$C.  If we were expecting this model to be useful, a deviation of $9^\circ$C is not particularly good.  However, considering the purpose of the project, it's not particularly bad.  When forecasting, it is important to define expectations.  If this model were to be relied upon to suggest, say, how warm to dress for each day in the forecast period, it would most likely be deemed a failure.  However, a model that suggests how warm to dress for a particular day would usually be expected to take into consideration many different aspects of the weather, only one of which might be temperature.  Instead of evaluating the model on whether it is personally practical one should use it to get a general idea of the temperature in the forecast period.  According to our predictions we see that temperature is trending up, into the positives, with an error metric of $9^\circ$C.  Given that March is roughly the beginning of a transition period between seasons (winter to spring) temperature might be particularly volatile.  This prediction is to be expected within the forecast period and is inline with conventional wisdom about temperature at that time of year.  Considering this and the fact that it predicts the general upward trend quite well, we believe the forecasts are acceptable, although there is certainly room for improvement. 

The most important factor to increased performance is data.  In this project we only consider past temperature as an indicator of future temperature (as well as a deterministic trend).  However, we know that temperature is the product of many natural phenomena occurring simultaneously and is a very complex system.  As such, effectively modelling it requires data that thoroughly captures the different mechanisms at play. Likewise, the chaotic nature of the weather makes it extremely difficult to forecast for long ranges as errors tend to compound rapidly.  While this challenge will be difficult to overcome it also presents an interesting opportunity for some very creative models for forecasting.  For example, U.S. orange juice production is concentrated in Central Florida and is very sensitive to weather fluctuations, specifically cold weather.  At the same time, futures contracts (agreements between two parties to buy/sell a product at a specified price at some future date) on orange juice are regularly traded on a public exchange (NYSE) where prices are subject to market forces.  Since the price of orange juice futures are a measure of information surrounding orange production in Central Florida and orange production is directly affected by the weather in Central Florida, it could be hypothesized that the price of OJ futures contains information about the weather.  In fact, there appears to be evidence for this hypothesis in which researchers have found a statistically significant relationship between the price of orange juice contracts and the weather in Central Florida$^5$.  This is a very interesting case study in the prediction of temperature and shows just how complex weather systems may be. 

For our analysis, the important thing to remember is the purpose of the study is an exercise in model specification.  It provides a good general description of the temperature within the forecast period and should not be confused for a more advanced result.  In spite of this expectation, it is still reassuring to see model performance in line with conventional wisdom, and serves as a good starting point in weather forecasting.

## Model Diagnostics

```r
candidates <- list(c(4,4), c(3,3), c(5,2), c(5,4), c(3, 2), c(4,2))
for (o in candidates){
    model <- arima(noise, order=c(o[1], 0, o[2]), method="ML", include.mean = TRUE, optim.control = list(maxit=1000))
    s <- 'Model: ARMA({o[1]},{o[2]})'
    
    model.res <- residuals(model)
    qqnorm(model.res, plot.it=TRUE, main=c('Normal QQ Plot: ', glue(s)))
    qqline(model.res)
    print(glue(s))
    print(shapiro.test(model.res))
    tsdiag(model)
}
```
