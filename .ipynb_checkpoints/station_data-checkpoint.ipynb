```r
library('dplyr')
library('devtools')
library('sf')
library('mapview')
```

```r
stations <- read.csv('Station_Inventory_EN.csv', skip=2, header=TRUE)
head(stations)
```

```r
TO_stations <- dplyr::filter(stations, grepl('TORONTO', Name))
station_coord <- select(TO_stations, Name, Longitude..Decimal.Degrees., Latitude..Decimal.Degrees.)
names(station_coord) <- c('Station', 'Longitude', 'Latitude')
station_coord <- as_tibble(station_coord)
locations_sf <- st_as_sf(station_coord, coords = c("Longitude", "Latitude"), crs = 4326)
```

```r
mapview(locations_sf)
```
